# Linear Regression

## Regression

It is a method used to determine how a dependent variable is related to one or more independent variables.

## Regression Analysis

Regression Analysis attempts to calculate the estimated value of the dependent variable given the values of the independent variables.

## Simple linear regression

Simple linear regression is a regression analysis used to model the relationship between one dependent variable and only one independent variable. In this analysis, the dependent variable is usually denoted as Y, while the independent variable is denoted as X, and the relationship between them is expressed by the formula Y = a + bX.

The main purpose of simple linear regression is to find the best line that describes the relationship between the values of the independent variable and the values of the dependent variable.

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
# Data set
data = {'area': [50, 70, 90, 110, 130],
        'price': [3000, 4000, 5000, 6000, 7000]}
df = pd.DataFrame(data)
```

```python
plt.xlabel('Area', color='blue')
plt.ylabel('Price', color='blue')
plt.title("Linear Regression", color='red')
plt.scatter(df.area, df.price, color='red', marker='*')
```

```python
X = df['area'].values.reshape(-1, 1)  
y = df['price'].values 
```

```python
model = LinearRegression()  # Create linear regression model
model.fit(df[['area']], y)  # Train the model
```

```python
model.coef_
```

```python
model.intercept_
```

```python
# Our linear regression formula:
# y = a + bX 

a = model.intercept_
b = model.coef_
x = 127
y = a + b * x
y
```

```python
plt.xlabel('Area', color='red')
plt.ylabel('Price', color='red')
plt.title("Linear Regression", color='red')
plt.scatter(df.area, df.price, color='red', marker='+')
plt.plot(df.area ,model.predict(df[['area']]), color = 'blue')
```
